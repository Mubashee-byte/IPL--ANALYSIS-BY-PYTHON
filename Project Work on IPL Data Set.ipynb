#  2: Check for Missing Values
##  Find missing values in a dataset.

```python
import pandas as pd  # Importing the pandas library for data handling

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

df.isnull().sum()  # Checking for missing values in each column and counting them
```

# 3 : Count Unique Values in a Column
## Count unique values in a specific column.

```python
import pandas as pd  # Importing the pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

un = (df['runs'].nunique())  # Counting the number of unique values in the 'runs' column

display('unique value of runs:', un)  # Displaying the count of unique run values
```

#  4: Group Data and Count
## Count occurrences of each unique value in a column.

```python
import pandas as pd  # Importing the pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

dc = (df["wickets"].value_counts())  # Counting occurrences of each unique value in the 'wickets' column

display("unique value of wickets:", dc)  # Displaying the count of unique wicket values
```

#  5: Filter Data Based on Condition
## Extract rows where a column value meets a condition.

```python
import pandas as pd  # Importing the pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

fc = df[df['runs'] > 30]  # Filtering rows where 'runs' column values are greater than 30

display(fc)  # Displaying the filtered DataFrame
```

#  7: Find the Average of a Column
##  Compute the average value of a numeric column.

```python
import pandas as pd  # Importing pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

av_scr = df['runs'].mean()  # Calculating the average (mean) of the 'runs' column

display("average of score:", av_scr.round(2))  # Displaying the rounded average score
```

# 8: Find the sum of a Column
## Compute the sum value of a numeric column.

```python
import pandas as pd  # Importing pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

sv = df['runs'].sum()  # Calculating the total sum of the 'runs' column

display("sum of score:", sv)  # Displaying the total sum of scores
```

# 9: Find Maximum & Minimum Values
##  Get the highest and lowest values in a column.

```python
import pandas as pd  # Importing pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

max_w = df['wickets'].max()  # Finding the maximum value in the 'wickets' column
min_w = df['wickets'].min()  # Finding the minimum value in the 'wickets' column

print('Maximum wickets in a match:\n', max_w)  # Printing the maximum wickets

print()  # Printing an empty line for better readability

print('Minimum wickets in a match:\n', min_w)  # Printing the minimum_
```

# 10 : Sort Data in Ascending/Descending Order
## Sort the dataset by a specific column.




```python
import pandas as pd  # Importing pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

sorted_df = df.sort_values("runs", ascending=False)  # Sorting data by 'runs' in descending order

display(sorted_df.head())  # Displaying the top 5 rows after sorting

```

```python
import pandas as pd  # Importing pandas library

df = pd.read_csv("ipl data.csv")  # Reading the CSV file into a DataFrame

ad = df.sort_values("runs", ascending=True)  # Sorting data by 'runs' in ascending order

display(ad.head())  # Displaying the top 5 rows after sorting
```
